Resident Population:
The resident population of the United States includes persons resident in the 50 States and the District of Columbia. An area's resident population consists of those persons "usually resident" in that particular area (where they live and sleep most of the time). The resident population excludes people whose usual residence is outside of the United States, such as the U.S. military and federal civilian personnel living overseas (and their dependents living with them), as well as private U.S. citizens living overseas. The resident population also excludes residents of the Commonwealth of Puerto Rico, and residents of the island areas under United States sovereignty or jurisdiction (principally American Samoa, Guam, Virgin Islands of the United States, and the Commonwealth of the Northern Mariana Islands).


Civilian Noninstitutionalized Population: 
All U.S. civilians not residing in institutional group quarters facilities such as correctional institutions, juvenile facilities, skilled nursing facilities, and other long-term care living arrangements.


Civilian Population:
All U.S. residents not in the active-duty military.


Household Population:
All U.S. residents who live in housing units such as single family homes, townhouses, apartments, and mobile homes.

```python
df=pd.read_csv("census_nationwide.csv")
```

```python
df.dtypes
```

```python
df["date"]=pd.to_datetime(df['date'])
#change date to date time format
```

```python
df
```

```python
#making all of the values from string to integer
df['Resident_Population']=df['Resident_Population'].astype(str).astype(int)
df['Civilian_Population']=df['Civilian_Population'].astype(str).astype(int)
df['Civilian_Noninstitutionalized_Population']=df['Civilian_Noninstitutionalized_Population'].astype(str).astype(int)
df['Household_Population']=df['Household_Population'].astype(str).astype(int)
```

```python
df
```

```python
df_nics=pd.read_csv('nics.csv')
```

```python
df_nics
```

```python
df_nics=df_nics.drop(["Unnamed: 0"], axis=1)
df_nics["month"]=pd.to_datetime(df_nics['month'])
#make month column into date time format
```

```python
df_nics.groupby(["year_only"])['total'].sum()
```

```python
df_nics_totals=df_nics.groupby(["month"])['total'].sum().reset_index()
df_nics_totals=df_nics_totals.rename(columns={'month': 'date'})
```

```python
df_pop=df[['date','Civilian_Noninstitutionalized_Population']]
#pull out the two columns im using for census
```

```python
#mgerging the NICS data with the census data
merged=pd.merge(df_nics_totals, df_pop, how="outer", on="date")
```

```python
merged
```

```python
#dividing total by population
merged['normalized'] = merged[['total']].div(merged['Civilian_Noninstitutionalized_Population'].values,axis=0)
```

```python
#multiple by 100,000 to get the per 100,000 values
merged['normalized']=merged['normalized']*100000
```

```python
merged
```

```python
merged.to_csv(r'nics_normal.csv')
```

```python
merged["month"] = merged['date'].dt.strftime('%Y-%m')
```

```python
df_data=pd.read_csv("alldata1.csv")
```

```python
df_data=df_data.drop(["Unnamed: 0"], axis=1)
df_data["date"]=pd.to_datetime(df_data['date'])
```

```python
df_data
```

```python
df_data["month"] = df_data['date'].dt.strftime('%Y-%m')
```

```python
df_data
```

```python
killed_month=df_data.groupby(["month"])["n_killed"].sum().reset_index()
```

```python
injured_month=df_data.groupby(["month"])["n_injured"].sum().reset_index()
```

```python
merged1=pd.merge(killed_month, df_pop, how="outer", on="month")
```

```python
merged2=pd.merge(injured_month, merged1, how="outer", on="month")
```

```python
merged2
```

```python
merged2['killed_normalized'] = merged2[['n_killed']].div(merged2['Civilian_Noninstitutionalized_Population'].values,axis=0)
merged2['killed_normalized']=merged2['killed_normalized']*100000
#calculating NICS per 100,000
```

```python
merged2['injured_normalized'] = merged2[['n_injured']].div(merged2['Civilian_Noninstitutionalized_Population'].values,axis=0)
merged2['injured_normalized']=merged2['injured_normalized']*100000
#calculating NICS per 100,000
```

```python
#merged all tables to include date, injured, killed, population, killed per 100,000 and injured per 100,000
merged2
```

```python
merged.plot(kind='line',x='month',y='normalized',color='green',figsize=(12,10))
#graph below is the NICS per 100,000
```

```python
merged2.plot(kind='line',x='month',y='killed_normalized',color='red',figsize=(12,10))
#graph is killed per 100,000
```

```python
merged2.plot(kind='line',x='month',y='injured_normalized',color='blue',figsize=(12,10))
#graph is injured per 100,000 
```

```python
merged3=pd.merge(merged, merged2, how="outer", on="date")
#merge killed and injured per 100,000 into one table
```

```python
merged3=merged3.drop({"Civilian_Noninstitutionalized_Population_x", "month_x", "month_y"}, axis=1)
```
