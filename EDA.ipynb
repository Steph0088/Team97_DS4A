```python
def yearly(df):

    df["date"] = pd.to_datetime(df["date"])
    monthly_accidents = df.groupby(df["date"].dt.to_period(freq ="Y")).size()
    
    return monthly_accidents
```

```python
yearly(df)
```

```python
def monthly(df):
    df["date"] = pd.to_datetime(df["date"])
    monthly_accidents = df.groupby(df["date"].dt.to_period(freq ="M")).size()
    
    return monthly_accidents
```

```python
monthly(df)
```

```python
yearly(df).plot(title="Gun Deaths per Year",
                figsize=(15,10));
```

```python
monthly(df).plot(title="Gun Deaths per Month",
                 color='red',
                 figsize=(15,10));
```
