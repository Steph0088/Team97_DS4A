```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.formula.api as sm
```

```python
df_vote = pd.read_csv('RedBlue1.csv')
df_vote
```

```python
df_vote_avg=df_vote[["state","AVG"]]
#only taking state and "AVG" for the vote status
```

```python
df_policy = pd.read_csv('gunpolicydata.csv')
```

```python
df_policy
```

```python
df_policytotal=df_policy[["state","year","lawtotal"]]
df_policytotal
#only keep state, year, law total
```

```python
df_policytotal=df_policy[["state","year","lawtotal"]]
df_policytotal=df_policytotal[df_policytotal.year > 2012]
#filtering to keep data from 2013-2020
```

```python
df_policytotal
```

```python
df_policytotal[(df_policytotal['state']) == "Pennsylvania"]
#confirming correct years were pulled
```

```python
poop=pd.merge(df_policytotal, df_vote_avg, how="outer",on="state")
poop #merge the two dataframes
```

```python
poop[(poop['AVG'] == "R") & (poop['year'] == 2013) ]
```

```python
law_byvote=poop.groupby(["AVG","year"])["lawtotal"].sum().reset_index()
law_byvote
```

```python
df = pd.read_csv('stateyear_injuredkilled.csv')
df=df.drop(["Unnamed: 0"],axis=1)
```

```python
df
```

```python
loop=pd.merge(df, df_vote_avg[["state","AVG"]], on=["state"])
loop
```

```python
law_byvote.to_csv(r'lawtotalRB.csv')
```

```python
injured_byvote=loop.groupby(["AVG","year"])['n_injured'].sum().reset_index()
injured_byvote
#table of states/vote, year, and number of people killed
```

```python
killed_byvote=loop.groupby(["AVG","year"])['n_killed'].sum().reset_index()
killed_byvote 
#table of states/vote, year, and number of people killed
```

```python
df_1 = killed_byvote.iloc[:8,:]
df_1
#table of only Democratic states and number of people killed
```

```python
df_2 = killed_byvote.iloc[8:,:]
df_2
#table of only Republican states and number of people killed
```

```python
df_1.plot(kind='line',x='year',y='n_killed',color='blue')
df_2.plot(kind='line',x='year',y='n_killed',color='red')
plt.show()
#graphed number killed, blue is democratic, red is republican
```

```python
df_3 = injured_byvote.iloc[:8,:]
df_3
#table of only Democratic states and number of people injured
```

```python
df_4 = injured_byvote.iloc[8:,:]
df_4
#table of only Democratic states and number of people injured
```

```python
df_3.plot(kind='line',x='year',y='n_injured',color='blue')
df_4.plot(kind='line',x='year',y='n_injured',color='red')
plt.show()
#graphed number injured, blue is democratic, red is republican
```

```python
df_5=law_byvote.iloc[:8,:]
df_5
#table of only Democratic states and number of laws
```

```python
df_6=law_byvote.iloc[8:,:]
df_6
#table of only Republican states and number of laws
```

```python
df_5.plot(kind='line',x='year',y='lawtotal',color='blue')
df_6.plot(kind='line',x='year',y='lawtotal',color='red')
plt.show()
#graphed number of laws, blue is democratic, red is republican
```
